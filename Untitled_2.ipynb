## Kita perlu merubah isi dari kolom Credit_History_Age dari format string ke angka

```python
data['Credit_History_Age']
```

```python
def convert_to_months(text):
    a = text.split(' ')
    years = int(a[0])
    months = int(a[3])
    return years*12+months

data['Credit_History_Age'] = data['Credit_History_Age'].apply(lambda x : x if pd.isna(x) else convert_to_months(x))
```

```python
X.dtypes
```

## Sebelum mulai proses training, mari kita encode non numeric data dengan OneHotEncoder dan menstandardkan data numeric dengan StandardScaler

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
# Identify categorical features
categorical_features = X.select_dtypes(include=['object', 'category']).columns
print(categorical_features)

scaler = StandardScaler()
scaled_x = pd.DataFrame(scaler.fit_transform(X.iloc[:,:-3]))
scaled_x.columns = scaled_x.columns.astype(str)

# Create an instance of the OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Encode the categorical features
encoded_df = pd.DataFrame(encoder.fit_transform(X[categorical_features]))
encoded_df.columns = encoded_df.columns.astype(str)
# Concatenate the encoded features with the remaining numerical features
#df_encoded = pd.concat([X.drop(columns=categorical_features).reset_index(), encoded_df.reset_index()], axis=True).drop(columns='index')
df_encoded = pd.concat([scaled_x.reset_index(), encoded_df.reset_index()], axis=True).drop(columns='index')
```

```python
df_encoded
```

## Split data ke test dan training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_encoded, y,test_size=0.20)
```

```python
models = pd.DataFrame(index=['train_accuracy', 'test_accuracy'],
                      columns=['KNN', 'RandomForest', 'Boosting'])
```

## Implementasikan KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X=X_test)
print(y_pred)
test_accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, knn.predict(X_train))
print(train_accuracy)
print(test_accuracy)
models.loc['train_accuracy','KNN'] = train_accuracy
models.loc['test_accuracy','KNN'] = test_accuracy
```

## Implementasikan RandomForest

```python
from sklearn.ensemble import RandomForestClassifier


RF = RandomForestClassifier(n_estimators=50, max_depth=50, random_state=55)
RF.fit(X_train, y_train)
y_pred = RF.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, RF.predict(X_train))
print(train_accuracy)
print(test_accuracy)
models.loc['train_accuracy','RandomForest'] = train_accuracy
models.loc['test_accuracy','RandomForest'] = test_accuracy
```

## Implementasikan AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier

boosting = AdaBoostClassifier(n_estimators=80, learning_rate=0.9, random_state=55)
boosting.fit(X_train, y_train)
y_pred = boosting.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
train_accuracy = accuracy_score(y_train, boosting.predict(X_train))
print(train_accuracy)
print(test_accuracy)
models.loc['train_accuracy','Boosting'] = train_accuracy
models.loc['test_accuracy','Boosting'] = test_accuracy
```
